```python
import pandas as pd
```

```python
url = "https://www.nrcs.usda.gov/wps/portal/nrcs/detail/national/home/?cid=nrcs143_013697"
df = pd.read_html(url)[0]
```

```python
abbvs = abbvs.rename(columns={'Full Name':'State'})
```
